```python
#Impor the data set using pandas
import pandas as pd
data= pd.read_csv("insurance_pre.csv")
```

```python
#Somker column categrical so convert to numeric using One Hot Encode
data= pd.get_dummies(data,drop_first='True')
```

```python
#Split input & Output variable
data.columns

independent= data[['age', 'bmi', 'children', 'sex_male', 'smoker_yes']]
dependent= data[[ 'charges']]
```

```python
#Split the x_train,train
#from sklearn.model_selection import train_test_split
#x_train,x_test,y_train,y_test= train_test_split(independent,dependent, test_size=0.30,random_state=0)
```

```python
#Standardize the model 
from sklearn.preprocessing import StandardScaler
sc= StandardScaler()
independent= sc.fit_transform(independent)
#x_train= sc.fit_transform(x_train)
#x_test= sc.transform(x_test)
independent
```

```python
from sklearn.svm import SVR
from sklearn.model_selection import GridSearchCV
param_grid = {'kernel':['poly', 'rbf','sigmoid','linear'],
              'C':[10,100,1000,2000,3000],'gamma':['auto','scale']}

grid= GridSearchCV(SVR(),param_grid,refit='True',verbose=3,n_jobs=-1)
grid.fit(independent,dependent)
```

```python
re= grid.cv_results_
print("The r2 value of the best parameter {}:".format(grid.best_params_))
print("The r2 value of the best score {}:".format(grid.best_score_))
```

```python
table= pd.DataFrame.from_dict(re)
```
