```python
#import the data set using pandas
import pandas as pd
data= pd.read_csv("insurance_pre.csv")
```

```python
#Update the Catogrical value
data= pd.get_dummies(data,drop_first='True')
```

```python
#Split the input & Output
data.columns
independent= data[['age', 'bmi', 'children', 'sex_male', 'smoker_yes']]
dependent= data[['charges']]
```

```python
#Split the train & test set
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test= train_test_split(independent,dependent,test_size=1/3,random_state=0)
```

```python
#Standardization to make the better performace
#from sklearn.preprocessing import StandardScaler
#sc= StandardScaler()
#x_train= sc.fit_transform(x_train)
#x_test= sc.transform(x_test)
```

```python
#Model Cretation
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestRegressor
param_grid={'criterion':['mse','mae'],'max_features': ['auto','sqrt','log2'],'n_estimators':[10,100]}

grid= GridSearchCV(RandomForestRegressor(),param_grid,refit='True',verbose=3,n_jobs=-1)
grid.fit(x_train,y_train)
```

```python
pred= grid.predict(x_test)
```

```python
print("The r2 value of the best Parameter:{}".format(grid.best_params_))
print("The r2 value of the best Score:{}".format(grid.best_score_))
```
